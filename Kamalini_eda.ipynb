```python
import matplotlib.pyplot as plt
import pandas as pd 
```

### Step 1: Read CSV file by panda

```python
df = pd.read_csv('carcrashes.csv')
print(df.head())
```

### Step 2: Initial data inspection

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

### Step 3: Initial descriptive statistics

```python
print(df.describe())
```

### Step 4: Initial Data Distribution for Numerical Columns

```python

# Inspect histogram by numerical column
#df['alcohol'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

### step 5: See the distribution between alcohol and number of non distraction 

```python

plt.scatter(df['alcohol'], df['not_distracted'])

# Add labels and title
plt.xlabel('Alcohol')
plt.ylabel('Not Distracted')
plt.title('Scatter Plot')
plt.plot()
plt.show()
```
